```python
dummy1 = pd.get_dummies(df1.Chest_pain)
dummy2 = pd.get_dummies(df1.Thalassemia_types)
dummy3 = pd.get_dummies(df1.ECG_results)
dummy4 = pd.get_dummies(df1.ST_slope)
dummy5 = pd.get_dummies(df1.Major_vessels)
merge = pd.concat([df1,dummy1,dummy2,dummy3,dummy4,dummy5],axis = 'columns')
```

```python
final = merge.drop(['Chest_pain','Thalassemia_types','ECG_results','ST_slope','Major_vessels'],axis = 1)
final.head()
```

```python
x = final.drop(['Heart_attack'], axis = 1)
y = final['Heart_attack']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.20, random_state = 5)
```

```python
feature_scaler = MinMaxScaler()
x_train = feature_scaler.fit_transform(x_train)
x_test = feature_scaler.transform(x_test)
```
